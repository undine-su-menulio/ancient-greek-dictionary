## Добро пожаловать! ##

Эта программа поможет вам вести ваш древнегреческо-русский словарик :)

Записывайте слова с переводом, просматривайте словарь с сортировкой по алфавиту и по частям речи, скачивайте в удобном формате с удобными настройками (идеально для Quizlet!), ищите контекст слов в корпусе папирусов, визуально следите за статистикой словаря на диаграммах и полюбуйтесь на красочное облако слов вашего словаря!

### Как сохранять словарик ###

**Периодически выгружайте словарик**, чтобы не потерять данные.

В Jupyter Notebook файлы сохраняются в [репозиторий](http://localhost:8888/tree), а в Google Colab - в файлы (Files) в меню слева (иконка папки). 

Обратите внимание: выпадающее File на панели управления - это не то, что вам нужно! Вам нужны именно Files в левом меню.

**Меню в Google Colab (слева)**:
* Левый скрин - само меню, отображающееся в левой верхней части экрана под панелью управления
* Правый верхний скрин - интерактивное Оглавление, открывающееся при клике на верхнюю иконку в меню
* Правый нижний скрин - Файлы (Files), куда загружаются и выгружаются файлы, с которыми вы работаете.

## Что делает эта программа ##

Что здесь можно делать?

1. **Создать словарик с нуля или загрузить его из файла в форматах .txt, .csv, .xls, .xlsx**
    1. Если вы хотите использовать ваш словарь, предварительно убедитесь, что он загружен в репозиторий Jupyter Notebook или в файлы (Files в выпадающем меню слева) в Google Colab. Новый файл можно загрузить там же с помощью кнопки Upload в правом верхнем углу.
    2. Словарик, предлагаемый для использования по умолчанию, называется dict_experimental.txt. Его структура - "словарь словарей", то есть {a: { b:c, d:e }, f: { g:h, i:j }}, где a, f - это части речи; b, d, g, i - это слова на греческом; c, e, h, j - это переводы на русский язык. Можно загружать и другие файлы такого формата, а также сохранять в таком формате ваши словарики. 
    
    3. Также можно загружать файлы форматов .txt и .csv, где части речи, слова на греческом и переводы перечислены через точку с запятой (;), причём каждое новое слово на новой строке, по типу:
        * глагол; θαυμάζω; восхищаться
        * прилагательное; μαρμαρόεις; сверкающий, мраморный     
        
    4. Словари из Excel должны содержать в первом столбце части речи (как в предлагаемом датасете), во втором - слова на греческом, в третьем - переводы на русский
        

2. **Вносить новые слова** в словарик, **удалять** ненужные слова и **изменять** записи в словарике

```python
# Ячейка №3

def uniformity (a):
    
    ''' Функция унифицирует греческий текст, избавляя его от надстрочных знаков и делая заглавные буквы - строчными. 
    Используется: а) чтобы Python правильно сортировал слова по алфавиту 
    (по умолчанию он думает, что буква с надстрочным знаком - это не та же самая, что без него), 
    б) чтобы изменение надстрочных знаков при склонении не влияло на распознавание текста,
    в) также спасает от случайного неуместного использования английских символов v, i, k, Y.
    На вход функция принимает текст на греческом языке.'''
    
    a = a.lower()
    a = re.sub ("ά|ἀ|ἁ|ἂ|ἃ|ἄ|ἅ|ἆ|ἇ|ὰ|ᾶ|ᾰ|ᾱ|ᾴ|ᾷ|ᾲ|ᾇ|ᾆ|ᾅ|ᾄ|ᾃ|ᾂ|ᾁ|ᾀ|ᾳ", 'α', a)
    a = re.sub ("έ|ἐ|ἑ|ἒ|ἓ|ἔ|ἕ|ὲ", 'ε', a)
    a = re.sub ("ή|ῃ|ᾐ|ᾑ|ᾒ|ᾓ|ᾔ|ᾕ|ᾖ|ᾗ|ῂ|ῇ|ῄ|ῆ|ὴ|ἧ|ἦ|ἥ|ἤ|ἣ|ἢ|ἡ|ἠ", 'η', a)
    a = re.sub ("ί|ἰ|ἱ|ἲ|ἳ|ἴ|ἵ|ἶ|ἷ|ὶ|ῖ|ῗ|ῒ|ΐ|ῐ|ῑ|ί|i", 'ι', a)
    a = re.sub ("ό|ὀ|ὁ|ὂ|ὃ|ὄ|ὅ|ὸ", 'ο', a)
    a = re.sub ("ύ|ὐ|ὑ|ὒ|ὓ|ὔ|ὕ|ὖ|ὗ|ὺ|ῦ|ῧ|ῢ|ΰ|ῠ|ῡ", 'υ', a)
    a = re.sub ('ώ|ὠ|ὡ|ὢ|ὣ|ὤ|ὥ|ὦ|ὧ|ὼ|ῶ|ῳ|ᾠ|ᾡ|ᾢ|ᾣ|ᾤ|ᾥ|ᾦ|ᾧ|ῲ|ῷ|ῴ', 'ω', a)
    a = re.sub ('ϰ|k', 'κ', a)
    a = re.sub ('v', 'ν', a)
    a = re.sub ('ῥ|ῤ', 'ρ', a)
    a = re.sub ('Ά|Ἀ|Ἁ|Ἂ|Ἃ|Ἄ|Ἅ|Ἆ|Ἇ|Ὰ|Ᾰ|Ᾱ|ᾏ|ᾎ|ᾍ|ᾌ|ᾋ|ᾊ|ᾉ|ᾈ|ᾼ|Ἀ', 'Α', a)
    a = re.sub ('Έ|Ἐ|Ἑ|Ἒ|Ἓ|Ἔ|Ἕ|Ὲ', 'Ε', a)
    a = re.sub ('Ή|Ἠ|Ἡ|Ἢ|Ἣ|Ἤ|Ἥ|Ἦ|Ἧ|Ὴ|ῌ|ᾘ|ᾙ|ᾚ|ᾛ|ᾜ|ᾝ|ᾞ|ᾟ', 'Η', a)
    a = re.sub ('Ί|Ἰ|Ἱ|Ἲ|Ἳ|Ἴ|Ἵ|Ἶ|Ἷ|Ὶ|Ῐ|Ῑ|I', 'Ι', a)
    a = re.sub ('Ό|Ὀ|Ὁ|Ὂ|Ὃ|Ὄ|Ὅ|Ὸ', 'Ο', a)
    a = re.sub ('Ύ|Ὑ|Ὓ|Ὕ|Ὗ|Ὺ|Ῠ|Ῡ|Y', 'Υ', a)
    a = re.sub ('Ώ|Ὠ|Ὡ|Ὢ|Ὣ|Ὤ|Ὥ|Ὦ|Ὧ|Ὼ|ᾯ|ᾮ|ᾭ|ᾬ|ᾫ|ᾪ|ᾩ|ᾨ|ῼ', 'Ω', a)
    return a
```

```python
# Ячейка №4

def uniformity_rus (a):
    
    ''' Функция принимает на вход текст на русском и унифицирует его, превращая "ё" в "е" 
    и делая заглавные буквы - строчными.  '''
    
    a = a.lower()
    a = re.sub ("ё", 'е', a)
    return a
```

```python
# Ячейка №5

def search_for_word (original_dic, name_of_column, mode, value):
    
    ''' Функция ищет слова или части слов в столбце pandas датафрейма. 
    Аргумент original_dic принимает на вход датафрейм pandas.
    Аргумент name_of_column принимает на вход название столбца, по которому нужно осуществить поиск. 
    Аргумент mode - либо 'contains', либо 'endswith'. 
    'contains' позволяет проверить, есть ли слово или часть слова в принципе. 
    'endswith' находит лишь ячейки, оканчивающиеся на нужное буквосочетание. 
    Аргумент value - либо None, либо любой другой текст (string). 
    Если value = None, то функция будет предлагать пользователю ввести буквосочетание, которое он хочет найти.
    Если value задано в функции, то будут возвращены результаты поиска по этому заданному значению.'''
    
    pandas_dic_2 = original_dic.copy()
    pandas_dic_2[name_of_column] = pandas_dic_2[name_of_column].apply (uniformity)
    if value == None:
        imported = input ('Введите слово, которое вы хотите найти: ')
        imported_pure = uniformity_rus (uniformity (imported)).strip()
    else:
        value_pure = uniformity_rus (uniformity (value))
    if mode == 'contains':
        if value == None:
            z = pandas_dic_2[pandas_dic_2[name_of_column].str.contains (imported_pure)]
        else:
            z = pandas_dic_2[pandas_dic_2[name_of_column].str.contains (value_pure)]
    elif mode == 'endswith':
        if value == None:
            z = pandas_dic_2[pandas_dic_2[name_of_column].str.endswith (imported_pure)]
        else:
            z = pandas_dic_2[pandas_dic_2[name_of_column].str.endswith (value_pure)]
    x = z.index.tolist()
    returned_dic = pandas_dic.iloc[x]
    pd.set_option ('display.max_rows', returned_dic.shape[0] + 1)
    return returned_dic
```

```python
# Ячейка №6

def df_to_dic (p_dataf, str_needed):
    
    ''' Функция принимает на вход pandas датафрейм и возвращает словарь словарей в формате { a: { b:c, d:e }, f: { g:h, i:j }} 
    В аргумент p_dataf вводится датафрейм pandas.
    Аргумент str_needed должен быть булевым числом (Boolean).
    Если str_needed = False, то вернётся обычный словарь (type dict).
    Если str_needed = True, то словарь вернётся в виде строки (type str) '''
    
    z = p_dataf.T.to_dict()
    we = []
    for a, b in z.items():
        qu = list (b.values())
        we.append (qu)
    re = {}
    for i in parts_of_speech:
        re.update ({i:{}})
    for i in we:
        re[i[0]].update ({i[1]:i[2]})
    if str_needed == False:
        return re
    else:
        wow_result = (str (re))
        return wow_result
```

```python
# Ячейка №7

def sort_ancient_gr (dictionary):
    
    ''' Функция принимает на вход pandas датафрейм и отсортировывает его сначала по частям речи 
    (согласно списку parts_of_speech), а затем по алфавиту по греческим словам, корректно обрабатывая надстрочные знаки. '''
    
    dictionary['help_column'] = dictionary['Слово на греческом'].apply (uniformity)
    dictionary['Часть речи'] = pd.Categorical (dictionary['Часть речи'], parts_of_speech)
    dictionary = dictionary.sort_values (by = ['Часть речи','help_column'])
    dictionary = dictionary.drop ('help_column',1).reset_index().drop ('index',1)
    return dictionary
```

```python
# Ячейка №9

def find_it_here (the_word_we_want):
    
    ''' Функция принимает на вход слово или часть слова на греческом (оно может не быть записано в словаре)
    и ищет его в корпусе египетских папирусов. Помогает понять контекст, в котором встречается данное буквосочетание
    в оригинальных текстах.'''
    
    url = 'http://papyri.info/search'
    urls = url + '?STRING=(' + the_word_we_want + ")&no_caps=on&no_marks=on&target=text&DATE_MODE=LOOSE&DOCS_PER_PAGE=15"
    page1 = requests.get (urls)
    soup1 = BeautifulSoup (page1.text)
    from_tags = soup1.findAll ('td',{'class':"kwic"})
    rubrics = []
    for i in from_tags:
        rubrics.append (i.text)
    for z in rubrics:
        print (z)
    if rubrics == []:
        print ('Увы! Ничего не найдено :(')
    if rubrics != []:
        save_ques = input ('Сохранить результат в файл .txt? Введите "Да" или "Нет": ')
        if save_ques == "Да" or save_ques == "да":
            rubrics_str = str (rubrics)
            papyri_name = input ('Введите желаемое название файла:').strip()
            with open (f'{papyri_name}.txt', 'w+', encoding='utf-8') as output:
                writer = csv.writer(output)
                writer.writerow (rubrics)
            print (f'Выгрузка файла "{papyri_name}.txt" прошла успешно!')
        else:
            print ('Результаты поиска не сохранены')
```

#### Количество слов по каждой части речи ####

Для отображения диаграммы, запустите ячейку ниже (*Ячейка №50*).

Скачать диаграмму на компьютер можно традиционно: правой кнопкой мыши по картинке - "Сохранить картинку как..."

```python
# Ячейка №50
# Столбиковая диаграмма "Количество слов по каждой части речи"

try:
    pofs_count = pd.DataFrame (pandas_dic['Часть речи'].value_counts()).reset_index()
    pofs_count['index'] = pofs_count['index'].astype ('str')
    x = pofs_count['Часть речи']
    y = pofs_count['index']

    fig = plt.figure (figsize = (25, 10))
    splot = sns.barplot (x, y)
    splot.axes.set_title ("Количество слов по каждой части речи", fontsize = 25)
    splot.set_xlabel ("Количество занесённых слов", fontsize = 20)
    splot.set_ylabel ("Часть речи", fontsize = 20)
    sns.set (font_scale = 1.25)

    totals = []
    for i in splot.patches:
        totals.append (int(i.get_width()))
    rects = splot.patches
    splot_height = 0
    for i in totals:
        splot.text (i + i/22, splot_height + 0.5, i,
                    ha = 'right', va = 'bottom')
        splot_height = splot_height + 1
    
    plt.savefig ('Количество слов по каждой части речи.png')
except NameError:
    print ('Вы либо не запустили Стартовые ячейки №1-9, либо не загрузили никакого словаря. Исправьте это и возвращайтесь сюда! :)')
```

#### Количество слов по роду и числу ####

Для отображения диаграммы, запустите ячейку ниже (*Ячейка №51*).

Скачать диаграмму на компьютер можно традиционно: правой кнопкой мыши по картинке - "Сохранить картинку как..."

Также диаграмма автоматически сохранится в репозиторий/ файлы как *'Количество слов по роду и числу.png'*.

```python
# Ячейка №51
# Столбиковая диаграмма "Количество слов по роду и числу"

try:
    male_sing_dic_count = search_for_word (pandas_dic, 'Слово на греческом', 'endswith', ' ὁ').shape[0]
    female_sing_dic_count = search_for_word (pandas_dic, 'Слово на греческом', 'endswith', ' ἡ').shape[0]
    neuter_sing_dic_count = search_for_word (pandas_dic, 'Слово на греческом', 'endswith', ' τό').shape[0]
    male_plur_dic_count = search_for_word (pandas_dic, 'Слово на греческом', 'endswith', ' οἱ').shape[0]
    female_plur_dic_count = search_for_word (pandas_dic, 'Слово на греческом', 'endswith', ' αἱ').shape[0]
    neuter_plur_dic_count = search_for_word (pandas_dic, 'Слово на греческом', 'endswith', ' τά').shape[0]

    genders = {'м.р. ед.ч.': male_sing_dic_count, 'ж.р. ед.ч.': female_sing_dic_count, 
           'ср.р. ед.ч.': neuter_sing_dic_count, 'м.р. мн.ч.': male_plur_dic_count, 
           'ж.р. мн.ч.': female_plur_dic_count, 'ср.р. мн.ч.': neuter_plur_dic_count}

    you = pd.DataFrame.from_dict (genders, orient = 'index').sort_values (by = 0, ascending = False)
    y_g = you.index
    x_g = you[0]

    fig2 = plt.figure (figsize = (12, 5))
    gen_plot = sns.barplot (x_g, y_g, palette = "Set2")
    gen_plot.axes.set_title ("Количество слов по роду и числу",fontsize = 15)
    gen_plot.set_xlabel ("Количество занесённых слов",fontsize = 13)
    gen_plot.set_ylabel ("Род и число",fontsize = 13)

    totals_g = []
    for i in gen_plot.patches:
        totals_g.append (int(i.get_width()))
    gen_height = 0
    for i in totals_g:
        gen_plot.text (i + i/17, gen_height + 0.06, i,
                       ha = 'right', va = 'bottom')
        gen_height = gen_height + 1
    
    plt.savefig ('Количество слов по роду и числу.png')
except NameError:
    print ('Вы либо не запустили Стартовые ячейки №1-9, либо не загрузили никакого словаря. Исправьте это и возвращайтесь сюда! :)')
```

2. Выберите из двух выпадающих меню (виджетов) цвет фона и язык слов для облака, которое вы хотите сформировать.

    2. Варианты языка слов: Слова на греческом или Переводы.
3. Запустите *Ячейку №53*.
4. Если вы хотите отобразить облако с другим фоном и/ или на другом языке, поменяйте значения в выпадающих меню (виджетах) в *Ячейке №52*.
